#### Numpy array properties

- ndim - the number of dimensions (axes) in the array 
- shape - the size of the array for each dimension 
- size - the total number of elements in the array 
- dtype - the data type of the elements in the array

```python
import numpy as np 
sales = [0,5,155,0,518,0,1827,616,317,325]

sales_array = np.array(sales)

type(sales_array)
```

```python
print(f"ndim: {sales_array.ndim}")
print(f"shape: {sales_array.shape}")
print(f"size: {sales_array.size}")
print(f"dtype: {sales_array.dtype}")
```

# Indexing & Slicing Arrays

Indexing & Slicing one-demensional arrays is the same as base python
- array[index] - indexing to access a single element (0-index)
- array[start:stop:step size] -  slicing to access a series of elements (stop is not inclusive)

```python
product_array = np.array(['fruits', 'vegetables', 'cereal', 'dairy', 'eggs', 'snacks', 'beverages', 'coffee', 'tea', 'spices'])

product_array
```

```python
product_array[1]
```

```python
product_array[-1]
```

```python
product_array[:5]
```

```python
product_array[5::2] # after 5 and step 2
```

Indexing & slicing two - dimensional arrays requires an extra index of slice
- array[row index, column indec] - indexing to access a single element (0-indexed)
- array[start:stop:step size, start:stop:step size] -  slicing to access a series of elements

```python
product_array = np.array(['fruits', 'vegetables', 'cereal', 'dairy', 'eggs', 'snacks', 'beverages', 'coffee', 'tea', 'spices'])

product_array2D = product_array.reshape(2,5)
product_array2D
```

```python
product_array2D[1,2]
```

```python
product_array2D[:,2:]
```

```python
product_array2D[1:,:]
```

# Assignment: Array Operations
![Screenshot 2026-07-05 at 23-24-57 Challenge Array operations LinkedIn Learning.png](<attachment:Screenshot 2026-07-05 at 23-24-57 Challenge Array operations LinkedIn Learning.png>)

```python
prices = np.array([5.99,6.99,22.49,99.99,4.99,49.99])
total = prices + 5
total
```

```python
import numpy as np

rng = np.random.default_rng(2022)

discount_pct = rng.random(6)

pct_owed = 1 - discount_pct

final_owed = total*pct_owed



print(discount_pct)
print(pct_owed)
print(final_owed.round(2))
```
